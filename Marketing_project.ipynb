```python
from kmodes.kmodes import KModes
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
cost = []
K = range(1,11)
for num_clusters in list(K):
    kmode = KModes(n_clusters=num_clusters, init = "random", n_init = 5, verbose=0)
    kmode.fit_predict(data)
    cost.append(kmode.cost_)
    
plt.plot(K, cost, 'bx-')
plt.xlabel('No. of clusters')
plt.ylabel('Cost')
plt.title('Elbow Method For Optimal k')
plt.show()
```

## Biggest elbows after 5th and 8th
#### Very unstable due to the small number observations and random choosing of initial cluster centers

```python
kmode = KModes(n_clusters=3, init = "random", n_init = 5,)
clusters = kmode.fit_predict(data)
clusters
```

```python
data.drop(columns='Cluster').insert(0, "Cluster", clusters, True)
data
```
